```python
import pandas as pd
movies = pd.read_csv('https://gist.githubusercontent.com/tiangechen/b68782efa49a16edaf07dc2cdaa855ea/raw/0c794a9717f18b094eabab2cd6a6b9a226903577/movies.csv')
movies
```

Visualizzare la somma dei Audience Score %

```python
movies["Audience score %"].sum()
```

Visualizzare la media dei Rottem Tomatoes %

```python
movies["Rotten Tomatoes %"].mean()
```

Visualizzare i nomi dei film che sono prodotti dalla universal

```python
movies[movies["Lead Studio"] == "Universal"][["Film"]]
```

Visualizzare i film e la casa produttrice che hanno un Rotten Tomatoes % superiore alla media

```python
movies[movies["Rotten Tomatoes %"] > movies["Rotten Tomatoes %"].mean()][["Film", "Lead Studio"]]
```

Visualizzare il film con l'audience score maggiore

```python
movies[movies["Audience score %"] == movies["Audience score %"].max()][["Film"]]
```

Visualizzare il minimo della colonna gross

```python
movies["Worldwide Gross"].min()
```
